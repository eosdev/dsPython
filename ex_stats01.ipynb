```python
import pandas as pd
import numpy as np
from pandas import read_excel
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
from scipy.stats import variation  
import numpy as np 
```

```python
#base de dados azulejo
azulejo = pd.read_csv('C:/Users/Eden/Documents/PUC-MG - Inteligência Artifical e Aprendizado de Máquina/EG/Azulejo_CDBD.csv', sep=';', dtype={'Espessura': float, 'Turma': str})
azulejo.head()
```

```python
azulejo.tail()
```

```python
#cria parâmetro de classificação
def mapeamento(Espessura):
    if 2<Espessura<3:
        return 'Espessura fina'
    elif 3>=Espessura<=4:
        return 'Espessura media'
    elif 4>Espessura<=5:
        return 'Espessura intermediaria'
    else:
        return 'Espessura alta'
```

```python
azulejo['Faixa'] = azulejo['Espessura'].apply(mapeamento)
```

```python
turmaA = azulejo[(azulejo.Turma == 'A')]
turmaB = azulejo[(azulejo.Turma == 'B')]
```

```python
azulejo_class = pd.crosstab(index=azulejo["Faixa"], 
                                columns=azulejo["Turma"],
                                margins=True)   # Include row and column totals

azulejo_class.columns = ["TurmaA","TurmaB","total"]
#azulejo_class.index= ["T1_A","T2_B","Total"]
```

```python
azulejo_class
```

```python
# To change parameters of density distribution
sns.distplot( a=azulejo["Espessura"],bins=20, kde=True,
kde_kws={"color": "r", "alpha":0.3, "linewidth": 5, "shade":True })
```

```python
#Estatísticas descritivas
print('Média: {}'.format(round(st.mean(azulejo['Espessura']),2)))
print('Moda: {}'.format(round(st.mode(azulejo['Espessura']),2)))
print('Mediana: {}'.format(round(st.median(azulejo['Espessura']),2)))
print('Variância: {}'.format(round(st.variance(azulejo['Espessura']),2)))
print('Desvio Padrão: {}'.format(round(st.stdev(azulejo['Espessura']),2)))
print('CV: {}'.format(round(variation(azulejo['Espessura']),2)))
```
