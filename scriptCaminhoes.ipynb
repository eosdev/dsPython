```python
#Declarando as variáveis
cards = []
card = {}

ff = webdriver.Firefox(executable_path="D:/Programas/gecko/geckodriver.exe")
for i in range(6): #75
    ff.get('https://www.vwtb.com.br/caminhoes/Delivery?id='+str(i+1))
    time.sleep(40)
    html = ff.page_source
    html2 = trata_html(html)
    soup = BeautifulSoup(html2,'html.parser')
    
    #Tags de interesse
    anuncios = soup.find('div', {'class': 'card-area row'}).find_all('div', {'class': 'truckCard card card-border'})
    anuncio = soup.find('div', {'class': 'truckCard card card-border'})
    
    #Coletando informações dos cartões
    for anuncio in anuncios:
        card = {}
    
        #Grupos de caminhões
        try:
            nomes = anuncio.find('h6', {'class': 'card-title subtitle-section'})
        except:
            nomes = ""

        names = []
        for nome in nomes:
            names.append(nome)
            card['nomes'] = names


        #Motor dos caminhões
        try:
            motores = anuncio.find_all('div', {'class': 'truckDetailedAttributes__value'})[0]
        except:
            motores = ""

        motors = []
        for motor in motores:
            motors.append(motor)
            card['motor'] = motors


        #Potencia dos caminhões
        try:
            potencias = anuncio.find_all('div', {'class': 'truckDetailedAttributes__value'})[1]
        except:
            potencias = ""

        power = []
        for potencia in potencias:
            power.append(potencia)
            card['potencia'] = power


        #Outras descricões
        try:
            descricoes = anuncio.find_all('div', {'class': 'truckDetailedAttributes__value'})[2]
        except:
            descricoes = ""

        detalhes = []
        for descricoe in descricoes:
            detalhes.append(descricoe)
            card['detalhes'] = detalhes


        #Trasmissão do caminhão
        try:
            transmissoes = anuncio.find_all('div', {'class': 'truckDetailedAttributes__value'})[3]
        except:
            transmissoes = ""

        transmissions = []
        for transmissao in transmissoes:
            transmissions.append(transmissao)
            card['trasmissao'] = transmissions



        #Peso do caminhão
        try:
            pesos = anuncio.find_all('div', {'class': 'truckDetailedAttributes__value'})[4]
        except:
            pesos = ""

        weight = []
        for peso in pesos:
            weight.append(peso)
            card['peso'] = weight


        #Obtendo imagens dos caminhões
        try:
            imagem = anuncio.find_all('img', {'class': 'card-img-top lazy loaded'})
        except:
            imagem = ""
        jpeg = []
        for image in imagem:
            jpeg.append(image.get('src'))
            card['link_imagem'] = jpeg
        
        #Salvando os arquivos e renomeando com sua respectiva imagem
        #y = image.get('src')
        picture_request = requests.get(jpeg[0])
    #if picture_request.status_code == 200:
        with open("C:/Users/edens/Downloads/caminhoes/"+nome.replace("/","")+".jpeg", 'wb') as f:
            f.write(picture_request.content)

    
    
    
        #Criando lista de cards
        cards.append(card)

ff.quit()    
    
#Criando o dataframe    
teste = pd.DataFrame(cards)
teste
```
