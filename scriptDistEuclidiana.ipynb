```python
#Importando a biblioteca matemática e gráfica em python
import math
import matplotlib.pyplot as plt
```

```python
#Função da biblioteca matemática do python parametrizadas com os pontos fornecidos
from scipy.spatial import distance
a = (1, 3)
b = (2, 5)
dst = distance.euclidean(a, b)
```

```python
#Resultado do cáculo da distância euclidiana
dst
```

```python
#Adicionando dados da distância euclidiana
line1, = plt.plot([1, 2, 3, 5], label=dst, linewidth=2)

# Cria uma legenda para a primeira linha
first_legend = plt.legend(handles=[line1], loc='upper right')

# Adicionando a legenda manualmente aos eixos atuais.
ax = plt.gca().add_artist(first_legend)
plt.title('Distância Euclidiana')
plt.show()
```
